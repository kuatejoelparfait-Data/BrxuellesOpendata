```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (13, 6)
plt.rcParams['font.family'] = 'DejaVu Sans'
print("Librairies chargees.")
```

## 1. Chargement des donnees

```python
# Amenagements cyclables
df_velo = pd.read_csv('amenagements_cyclables.csv', sep=';', encoding='utf-8-sig')
print("Velo — shape:", df_velo.shape)
print("Colonnes:", df_velo.columns.tolist())
df_velo.head()
```

```python
# Parking PMR
df_pmr = pd.read_csv('parking_pmr.csv', sep=';', encoding='utf-8-sig')
print("PMR — shape:", df_pmr.shape)
print("Colonnes:", df_pmr.columns.tolist())
df_pmr.head()
```

## 2. Analyse des amenagements cyclables

```python
# Types d'amenagements cyclables
type_col = [c for c in df_velo.columns if 'Type' in c and 'code' not in c.lower()][0]
statut_col = [c for c in df_velo.columns if 'Statut' in c and 'code' not in c.lower()][0] if any('Statut' in c and 'code' not in c.lower() for c in df_velo.columns) else None
commune_col = [c for c in df_velo.columns if 'Commune' in c][0]

print("Colonne type:", type_col)
print("Colonne statut:", statut_col)
print("Valeurs uniques type:", df_velo[type_col].value_counts().head(10))
```

```python
# Repartition par type d'amenagement cyclable
type_counts = df_velo[type_col].value_counts().head(12).sort_values(ascending=True)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Graphique 1 : Types d'amenagements
axes[0].barh(type_counts.index, type_counts.values, 
             color=sns.color_palette("Blues_r", len(type_counts)))
axes[0].set_title("Types d'amenagements cyclables a Bruxelles\n(Bruxelles Mobilite)", 
                  fontweight='bold', fontsize=12)
axes[0].set_xlabel("Nombre de troncons/entites")
for i, v in enumerate(type_counts.values):
    axes[0].text(v + 3, i, str(v), va='center', fontsize=8)

# Graphique 2 : Statut des amenagements
if statut_col:
    statut_counts = df_velo[statut_col].value_counts()
    axes[1].pie(statut_counts.values, labels=statut_counts.index, 
                autopct='%1.1f%%',
                colors=sns.color_palette("Pastel2", len(statut_counts)), 
                startangle=140)
    axes[1].set_title("Statut des amenagements cyclables\n(en service / en travaux / planifie)", 
                      fontweight='bold', fontsize=12)
else:
    axes[1].text(0.5, 0.5, 'Statut non disponible', ha='center', va='center')

plt.tight_layout()
plt.savefig('01_amenagements_cyclables_types.png', dpi=150, bbox_inches='tight')
plt.show()
print("Graphique sauvegarde.")
```

```python
# Amenagements cyclables par commune
commune_velo = df_velo[commune_col].value_counts().sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(13, max(7, len(commune_velo)*0.5)))
colors = sns.color_palette("Blues_r", len(commune_velo))
bars = ax.barh(commune_velo.index, commune_velo.values, color=colors)
ax.set_title("Nombre d'amenagements cyclables par commune\n(Ville de Bruxelles — Bruxelles Mobilite)", 
             fontweight='bold', fontsize=13)
ax.set_xlabel("Nombre de troncons / entites")
for bar in bars:
    w = bar.get_width()
    ax.text(w + 1, bar.get_y() + bar.get_height()/2, str(int(w)), va='center', fontsize=8)
plt.tight_layout()
plt.savefig('02_cyclable_par_commune.png', dpi=150, bbox_inches='tight')
plt.show()
print("Graphique sauvegarde.")
```

## 3. Analyse des places PMR

```python
# Types de places PMR et nombre de places
type_pmr_col = [c for c in df_pmr.columns if 'Type' in c][0] if any('Type' in c for c in df_pmr.columns) else None
places_col = [c for c in df_pmr.columns if 'Nombre' in c or 'places' in c.lower()][0] if any('Nombre' in c or 'places' in c.lower() for c in df_pmr.columns) else None

print("Col type PMR:", type_pmr_col)
print("Col places:", places_col)
if type_pmr_col:
    print(df_pmr[type_pmr_col].value_counts())
```

```python
# Analyse PMR
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

if type_pmr_col:
    pmr_type = df_pmr[type_pmr_col].value_counts()
    axes[0].pie(pmr_type.values, labels=pmr_type.index, autopct='%1.1f%%',
               colors=sns.color_palette("Set3", len(pmr_type)), startangle=90)
    axes[0].set_title("Types de places PMR\na Bruxelles", fontweight='bold')

if places_col:
    df_pmr_clean = df_pmr.copy()
    df_pmr_clean[places_col] = pd.to_numeric(
        df_pmr_clean[places_col].astype(str).str.replace(',', '.'), errors='coerce')
    total_places = df_pmr_clean[places_col].sum()
    avg_places = df_pmr_clean[places_col].mean()
    
    axes[1].hist(df_pmr_clean[places_col].dropna(), bins=15, 
                color='coral', edgecolor='white', alpha=0.85)
    axes[1].axvline(avg_places, color='darkred', linestyle='--', linewidth=2,
                   label=f'Moyenne: {avg_places:.1f} places')
    axes[1].set_title(f"Distribution du nombre de places PMR par site\nTotal: {int(total_places)} places", 
                     fontweight='bold')
    axes[1].set_xlabel("Nombre de places par site")
    axes[1].set_ylabel("Frequence")
    axes[1].legend()
else:
    nb_pmr = len(df_pmr)
    axes[1].text(0.5, 0.5, f'Total: {nb_pmr} sites PMR recensees', 
                ha='center', va='center', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.savefig('03_analyse_pmr.png', dpi=150, bbox_inches='tight')
plt.show()
print("Graphique sauvegarde.")
```

## 4. Croisement : Velo vs PMR — Bilan territorial

```python
# Comparaison globale : nombre de troncons velo vs sites PMR
total_velo = len(df_velo)
total_pmr = len(df_pmr)

# Si colonnes communes disponibles
pmr_zone_col = [c for c in df_pmr.columns if 'Zone' in c][0] if any('Zone' in c for c in df_pmr.columns) else None

fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# Comparaison volumes
categories = ['Amenagements cyclables\n(troncons)', 'Places PMR\n(sites)']
values = [total_velo, total_pmr]
colors = ['steelblue', 'coral']
bars = axes[0].bar(categories, values, color=colors, width=0.5, edgecolor='white', linewidth=1.5)
for bar, v in zip(bars, values):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20,
                str(v), ha='center', fontweight='bold', fontsize=14)
axes[0].set_title("Volume compare : amenagements cyclables\nvs sites PMR a Bruxelles", 
                  fontweight='bold', fontsize=13)
axes[0].set_ylabel("Nombre total d'entites recensees")

# Evolution temporelle des amenagements cyclables si date disponible
date_col = [c for c in df_velo.columns if 'date' in c.lower() or 'begin' in c.lower()][0] if any('date' in c.lower() or 'begin' in c.lower() for c in df_velo.columns) else None
if date_col:
    df_velo_dated = df_velo.copy()
    df_velo_dated['annee'] = pd.to_datetime(df_velo_dated[date_col], errors='coerce').dt.year
    evolution = df_velo_dated.groupby('annee').size().reset_index(name='count')
    evolution = evolution.dropna().query('annee >= 2000 and annee <= 2026')
    evolution['cumul'] = evolution['count'].cumsum()
    
    axes[1].fill_between(evolution['annee'], evolution['cumul'], alpha=0.7, color='steelblue')
    axes[1].plot(evolution['annee'], evolution['cumul'], color='steelblue', linewidth=2)
    axes[1].set_title("Evolution cumulative des amenagements\ncyclables a Bruxelles", 
                      fontweight='bold', fontsize=13)
    axes[1].set_xlabel("Annee")
    axes[1].set_ylabel("Nombre cumule d'amenagements")
    axes[1].xaxis.set_major_locator(mticker.MaxNLocator(integer=True))
else:
    # Repartition par type sous forme de donut
    type_top = df_velo[type_col].value_counts().head(6)
    wedges, texts, autotexts = axes[1].pie(type_top.values, labels=type_top.index, 
                                            autopct='%1.0f%%',
                                            colors=sns.color_palette("Blues_r", len(type_top)),
                                            startangle=90, pctdistance=0.85)
    centre_circle = plt.Circle((0,0), 0.70, fc='white')
    axes[1].add_patch(centre_circle)
    axes[1].set_title("Top 6 types d'amenagements cyclables", fontweight='bold', fontsize=13)

plt.tight_layout()
plt.savefig('04_comparaison_velo_pmr.png', dpi=150, bbox_inches='tight')
plt.show()
print("Graphique sauvegarde.")
```

## 5. Interpretation politique et conclusions

### Constats principaux

1. **Predominance des pistes cyclables** : Les amenagements cyclables a Bruxelles sont nombreux et diversifies (bandes cyclables, pistes separees, zones de partage), mais leur repartition reste inegale selon les communes.

2. **PMR : couverture geographique** : Les places de stationnement PMR sont recensees sur l'ensemble du territoire, mais la densite varie significativement selon les zones. Les quartiers commerciaux et administratifs sont mieux equipes.

3. **Tension mobilite douce vs accessibilite PMR** : La reconversion de places de parking en pistes cyclables cree une tension politique entre les partisans de la mobilite douce et les besoins des personnes a mobilite reduite.

4. **Croissance des amenagements velo** : L'evolution temporelle montre une acceleration des investissements cyclables, notamment depuis le Plan de Mobilite Good Move (2019).

### Question politique repondue

> L'espace public bruxellois favorise-t-il equitablement la mobilite douce et l'accessibilite PMR ?

**Reponse** : Les donnees suggerent un effort significatif sur les amenagements cyclables, mais une vigilance est necessaire pour garantir que cet effort ne se fasse pas au detriment de l'accessibilite des PMR. Une politique de mobilite inclusive doit integrer les deux dimensions.

### Recommandations politiques

- Cartographier les zones de tension entre pistes cyclables et places PMR pour anticiper les conflits d'usage.
- Creer des zones multimodales incluant stationnement PMR securise ET acces velo.
- Renforcer les amenagements cyclables dans les communes peripheriques actuellement sous-equipees.
